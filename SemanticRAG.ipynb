```python
import pandas as pd
import spacy
from deep_translator import GoogleTranslator
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

class SemanticTranslator:
    def __init__(self):
        self.nlp = spacy.load("en_core_web_sm")
        self.model = SentenceTransformer('all-MiniLM-L6-v2')

    def translate_long_text(self, text, chunk_size=5000):
        # First check if text is None or empty
        if pd.isnull(text) or not str(text).strip():
            print("Empty or null text found, skipping...")
            return ""  # Return empty string instead of the original text

        try:
            text_str = str(text)
            translated_text = []

            chunks_needed = len(text_str) // chunk_size + (1 if len(text_str) % chunk_size != 0 else 0)
            print("Total chunks needed: ", chunks_needed)

            # Only proceed if text is not empty
            if chunks_needed > 0:
                for i in range(chunks_needed):
                    start = i * chunk_size
                    end = start + chunk_size
                    chunk = text_str[start:end]
                    if chunk.strip():  # Only translate if chunk has content
                        translated_chunk = GoogleTranslator(source="bn", target="en").translate(chunk)
                        if translated_chunk:  # Only append if translation succeeded
                            translated_text.append(translated_chunk)

            return "".join(translated_text)

        except Exception as e:
            print(f"Translation error: {str(e)}")
            return ""  # Return empty string on error

    def semantic_chunk(self, translated_text, similarity_threshold=0.5):
        try:
            # Process the translated text with spaCy
            doc = self.nlp(translated_text)
            sentences = [sent.text.strip() for sent in doc.sents]

            if not sentences:
                print("No sentences found in the translated text")
                return []

            # Get sentence embeddings
            embeddings = self.model.encode(sentences)
            # Calculate similarity between consecutive sentences
            similarities = cosine_similarity(embeddings)

            # Create chunks based on semantic similarity
            chunks = []
            current_chunk = [sentences[0]]

            for i in range(1, len(sentences)):
                # If current sentence is semantically similar to previous sentence
                if similarities[i-1][i] > similarity_threshold:
                    current_chunk.append(sentences[i])
                else:
                    # If not similar, complete current chunk and start new one
                    chunks.append(" ".join(current_chunk))
                    current_chunk = [sentences[i]]

            # Add the final chunk
            if current_chunk:
                chunks.append(" ".join(current_chunk))

            return chunks

        except Exception as e:
            print(f"Error in semantic chunking: {str(e)}")
            return []

def process_dataframe(data):
    # Create translator instance
    translator = SemanticTranslator()

    # Process each row in the dataframe
    results = []

    for idx, row in data.iterrows():
        print(f"\nProcessing row {idx}")

        # Get text from post_text column
        text = row['post_text']

        print("Original Text:")
        print(text)
        print("\n" + "="*50 + "\n")

        # Step 1: Translate
        print("Translating...")
        translated = translator.translate_long_text(text)
        print("Translated Text:")
        print(translated)
        print("\n" + "="*50 + "\n")

        # Step 2: Create semantic chunks
        print("Creating semantic chunks...")
        chunks = translator.semantic_chunk(translated)
        print("Semantic Chunks:")
        for i, chunk in enumerate(chunks, 1):
            print(f"\nChunk {i}:")
            print(chunk)

        # Store results
        results.append({
            'original_text': text,
            'translated_text': translated,
            'semantic_chunks': chunks
        })

        print("\n" + "="*70 + "\n")

    # Add results to dataframe
    data['translated_text'] = [r['translated_text'] for r in results]
    data['semantic_chunks'] = [r['semantic_chunks'] for r in results]

    return data

# Use the code
if __name__ == "__main__":
    # Assuming you have your dataframe in 'data'
    # Process the dataframe
    processed_data = process_dataframe(data)

    # Show sample results
    print("\nSample of processed data:")
    print(processed_data[['post_text', 'translated_text', 'semantic_chunks']].head())
```
